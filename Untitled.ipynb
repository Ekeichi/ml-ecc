```python
import pandas as pd
import numpy as np
import yfinance as yf
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import HistGradientBoostingRegressor, RandomForestRegressor
from sklearn.feature_selection import RFE
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.base import BaseEstimator, TransformerMixin
from ta.trend import SMAIndicator, EMAIndicator, MACD
from ta.momentum import RSIIndicator, StochasticOscillator
from ta.volatility import BollingerBands, AverageTrueRange
from ta.volume import OnBalanceVolumeIndicator
```

```python
# Transformateur personnalisé pour aplatir les données
class FlattenTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.original_shape = None
    
    def fit(self, X, y=None):
        self.original_shape = X.shape
        return self
    
    def transform(self, X):
        return X.reshape(X.shape[0], -1)
    
    def inverse_transform(self, X):
        if self.original_shape is not None:
            return X.reshape(self.original_shape)
        return X
```

```python
# Option 1: Juste après le téléchargement avec yfinance
ticker = 'AAPL'
data = yf.download(ticker, start='2015-01-01', end='2024-12-31')
# Garde seulement les colonnes pour AAPL et supprime le niveau 'Ticker'
data = data.loc[:, pd.IndexSlice[:, ticker]] # Sélectionne seulement AAPL si plusieurs tickers étaient là
data.columns = data.columns.droplevel(1) # Supprime le niveau 'Ticker'

# Option 2: Si tu as déjà le MultiIndex
# data.columns = data.columns.droplevel(1) # Supprime le niveau 'Ticker'

# Maintenant, les noms de colonnes sont simples ('Close', 'High', etc.)
print(data.columns)

# Et tu peux calculer et utiliser 'Daily_Pct_Change' normalement
data['Daily_Pct_Change'] = data['Close'].pct_change() * 100
print(data[['Close', 'Daily_Pct_Change']].head())
data.dropna(subset=['Daily_Pct_Change'], inplace=True) # Ceci fonctionnera maintenant
```

```python
from ta.trend import SMAIndicator, EMAIndicator, MACD
from ta.momentum import RSIIndicator
from ta.volatility import BollingerBands

# Ajouter des indicateurs techniques
data['SMA_20'] = SMAIndicator(data['Close'], window=20).sma_indicator()  # Moyenne mobile simple
data['EMA_12'] = EMAIndicator(data['Close'], window=12).ema_indicator()  # Moyenne mobile exponentielle
data['RSI_14'] = RSIIndicator(data['Close'], window=14).rsi()  # Indice de force relative
macd = MACD(data['Close'])
data['MACD'] = macd.macd()
data['MACD_Signal'] = macd.macd_signal()
bb = BollingerBands(data['Close'], window=20)
data['BB_High'] = bb.bollinger_hband()
data['BB_Low'] = bb.bollinger_lband()
for lag in range(1, 4):
    data[f'Volume_lag_{lag}'] = data['Volume'].shift(lag)
data['Day_of_Week'] = data.index.dayofweek
data['Month'] = data.index.month
data['Day_of_Week_sin'] = np.sin(2 * np.pi * data['Day_of_Week'] / 7)
data['Day_of_Week_cos'] = np.cos(2 * np.pi * data['Day_of_Week'] / 7)
data['Month_sin'] = np.sin(2 * np.pi * data['Month'] / 12)
data['Month_cos'] = np.cos(2 * np.pi * data['Month'] / 12)
data['Is_Quarter_End'] = data.index.is_quarter_end.astype(int)
data['Volatility_20'] = data['Daily_Pct_Change'].rolling(window=20).std()
data['Normalized_Volatility'] = data['Volatility_20'] / data['Daily_Pct_Change'].rolling(window=252).std()
data['Close_Volume_Ratio'] = data['Close'] / data['Volume']
data['RSI_MACD_Interaction'] = data['RSI_14'] * data['MACD']
data['ATR_14'] = AverageTrueRange(data['High'], data['Low'], data['Close'], window=14).average_true_range()
data['OBV'] = OnBalanceVolumeIndicator(data['Close'], data['Volume']).on_balance_volume()
data['Stoch'] = StochasticOscillator(data['High'], data['Low'], data['Close'], window=14).stoch()
data['Sum_Pct_Change_5d'] = data['Daily_Pct_Change'].rolling(window=5).sum()  # Somme des variations sur 5 jours
data['Mean_Pct_Change_10d'] = data['Daily_Pct_Change'].rolling(window=10).mean()  # Moyenne sur 10 jours

# Télécharger S&P 500
sp500 = yf.download('^GSPC', start='2015-01-01', end='2024-12-31')['Close']
sp500 = sp500.reindex(data.index).fillna(method='ffill')
data['SP500_Pct_Change'] = sp500.pct_change() * 100
# Vérifier les nouvelles colonnes
print(data.columns)
```

```python
#Liste des features
features_list = [
    'Close', 'Volume', 'Daily_Pct_Change', 'SMA_20', 'EMA_12', 'RSI_14', 
    'MACD', 'MACD_Signal', 'BB_High', 'BB_Low', 'ATR_14', 'OBV', 'Stoch',
    'Volume_lag_1', 'Volume_lag_2', 'Volume_lag_3', 
    'Day_of_Week_sin', 'Day_of_Week_cos', 'Month_sin', 'Month_cos', 'Is_Quarter_End',
    'Volatility_20', 'Normalized_Volatility', 'Close_Volume_Ratio', 'RSI_MACD_Interaction',
    'Sum_Pct_Change_5d', 'Mean_Pct_Change_10d', 'SP500_Pct_Change'
]

# Imputer les NaN
data[features_list] = SimpleImputer(strategy='median').fit_transform(data[features_list])
data = data.dropna(subset=['Daily_Pct_Change'])
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestRegressor
import pandas as pd

```

```python
# Créer les séquences
window_size = 7
target_col = 'Daily_Pct_Change'

X_sequences = []
y_target = []

for i in range(window_size, len(data)):
    sequence = data[features_list].iloc[i-window_size:i].values
    target = data[target_col].iloc[i]
    X_sequences.append(sequence)
    y_target.append(target)

X_sequences = np.array(X_sequences)
y_target = np.array(y_target)
```

```python
# Diviser les données
split_ratio = 0.8
split_index = int(len(X_sequences) * split_ratio)
X_train = X_sequences[:split_index]
X_test = X_sequences[split_index:]
y_train = y_target[:split_index]
y_test = y_target[split_index:]
```

```python
# Définir la pipeline avec RFE
pipeline = Pipeline([
    ('flatten', FlattenTransformer()),  # Aplatir les données 3D
    ('imputer', SimpleImputer(strategy='median')),  # Imputer les NaN
    ('scaler', StandardScaler()),  # Mettre à l'échelle
    ('feature_selection', RFE(estimator=RandomForestRegressor(n_estimators=100, random_state=42), 
                             n_features_to_select=50)),  # RFE avec RandomForest
    ('model', HistGradientBoostingRegressor(random_state=42))  # Modèle final
])
```

```python
# Entraîner la pipeline
pipeline.fit(X_train, y_train)
```

```python
# Prédire
y_pred = pipeline.predict(X_test)
```

```python
# Évaluer
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
print(f"RMSE Pipeline (HGB avec RFE) : {rmse:.2f}")
print(f"MAE Pipeline (HGB avec RFE) : {mae:.2f}")
```

```python
# Identifier les features sélectionnées
selected_mask = pipeline.named_steps['feature_selection'].support_
feature_columns = [f"{feat}_t-{t}" for feat in features_list for t in range(window_size)]
selected_cols = [feature_columns[i] for i in range(len(feature_columns)) if selected_mask[i]]
selected_features = list(set([col.split('_t-')[0] for col in selected_cols]))
print(f"Features sélectionnées : {selected_features}")
```
